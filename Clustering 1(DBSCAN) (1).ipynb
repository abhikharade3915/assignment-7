```python
array=data_air1.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=6)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
DD=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
DD
```

```python
pd.concat([data_air1,DD],axis=1)
```
